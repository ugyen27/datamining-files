```python
import seaborn as sns

```

```python
# Loading Data

penguins = sns.load_dataset("penguins")

penguins

# type(penguins) 
```

```python
# Lineplot 

sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g")
```

This graph shows the correlation of body mass of penguin(gram) and their flipper length(mm). The graph shows a strong positive correlation that is those penguins who have higher body mass tends to have longer flipper.

```python
sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex")
```

This graph shows that male penguins have more body mass and longer flipper length than female penguins.

```python
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

This scatter plot shows the relation between penguins bill length and bill depth. The graph indicates that longer bill depth has nothing to do with deeper bill depth. Penguins can have longer bill length while  having shallow bill depth or peguins can have longer bill length with deep bill depth.

```python
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

This graph shows the penguins bill length and bill depth according to their species. From the graph we can see that Adelie have the deepest bill depth and Chinstrap have longest bill length among these three species. Gentoo has the lowest bill depth and Adelie have the shortest bill length.

```python
sns.displot(data=penguins, x="flipper_length_mm", col="island")
```

The graph shows the length of flipper of penguin according to their island. The graph shows that most penguins in Torgersen island have flipper length from 190 to 195mm. The penguins of Biscoe have flipper length of 215 to 220mm and penguins of Dream island have most flipper length between 190-195mm. we could also see that Biscoe penguin have longer flipper length than other two islands penguin.

```python
sns.histplot(data=penguins, x="body_mass_g")
```

This graph shows the body mass of every penguin. Most penguin have the body mass of 3750 to 4000g. While the heaviest body mass of the penguin werebetween 6000 to 6250g and least body mass were between 2750 to 3000 grams which are both rare.

```python
sns.pairplot(penguins)
```

These various scatter plot is of data of penguins. Most of the graph have positive correlation which indicate that when penguin have longer bill length, more bill depth or more body mass, there is a high chance that their other attributes will also will higher.

```python
sns.boxplot( penguins['bill_depth_mm'] ) 
```

The box and whisker shows the minimum bill depth whic is 12mm, maximum bill depth which is 22 and median which is between 17 and 18.

```python
sns.boxplot( x=penguins['species'], y=penguins['body_mass_g'] ) 
```

The box plot are of three species of penguins. From here, we can see that Gentoo species have the most highest body mass. We can also see that the penguins with highest body mass belongs to Gentoo species. The peguins which have the least body mass belongs to Adelie species.
