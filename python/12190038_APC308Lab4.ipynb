```python
# plotting from a dataset of 'flights'

fData = sns.load_dataset("flights")

fData
```

# A. Line Graph

```python
sns.lineplot(data=fData, x="passengers", y="year")
```

```python
sns.lineplot(data=fData, x="passengers", y="year", hue="month")
```

# This line graph shows that as the year increases , the number of passengers flying over the flight is also increasing from which the maximum is approximately more than 600 and Minimum is 100 passengers. The month that the passenger took the maximum flight is at the range of july to August.

# B. Scatter Plot

```python
sns.scatterplot(data=fData, x="passengers", y="month")
```

# To compare the number of passenger in each month meaning at what month the passengers are flying the most. So by looking at the graph i could figure out that at the mid of july there are more than 600 passengers which is actually the heavy month for the flight company whlist the more profit they earn the most in this month.In addition, at the November and Febuary month there are only less number of passenger taking the flight(300-400 numbers).

# C. Distribution Plotting

```python
sns.histplot(data=fData, x="passengers")  
```

# Most of time the number of passenger in flight is 200(28 flights) and the least is 600 passengers(3 flights).  

# D. Pair Plot

```python
sns.pairplot(flights, hue="month")
```

```python

import warnings                          # to silence warning if need be
warnings.filterwarnings('ignore')
sns.boxplot( flights['passengers'] )    
```

# The maximum passengers in flight is 600 and the minimum is 100 where the median is around 280. 

```python
sns.boxplot( x=flights['passengers'], y=flights['month'] )
```

# This graph shows the minimum ,maximum and median number of passengers according to each month. 

# F. Heat Map

```python
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights)
```

# The brightest color represents the maximum number of passengers.
