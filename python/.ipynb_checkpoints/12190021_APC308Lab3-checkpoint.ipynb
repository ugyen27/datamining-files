```python
#load the Applicants data set from the current directory

App = pd.read_csv('Applicants.csv')
App
```

```python
#Export to Excel 
App.to_excel("App.xlsx", sheet_name="App Excel Format")
```

```python
App.head()   
App.head(10)
App.tail()
App.tail(10)
```

```python
# some useful attributes  
App.dtypes

App.index

App.columns

App.info() 

App.T
```

```python
# Filtering

#Filtering columns by name
App['ApplicantIncome']

App.ApplicantIncome

App[ ['ApplicantIncome', 'LoanAmount']]

App['Loan_Amount_Term'] * App['Credit_History']
```

```python
#Filtering rows by slicing
App[0:20:2]

App[610:600:-2]

App.loc[:]

#Filtering columns by label: loc
App.loc[:20:3, ["Loan_ID", "ApplicantIncome"]]

#Filtering by values
App[App['LoanAmount'] > 500]

App[(App['ApplicantIncome'] > 0) & (App['LoanAmount'] > 500)]
```

```python
#Aggregation / Summary Stats

#App.mean()

#App.mean(0)

App["LoanAmount"].mean()

App["LoanAmount"].std()

App["LoanAmount"].min()

App["LoanAmount"].max()

App["LoanAmount"].median()
App["LoanAmount"].describe()

App["Loan_Amount_Term"].value_counts()
```

```python
# Sorting and copying

App.sort_values(by="Loan_ID")


App.sort_values(by="LoanAmount")


App.sort_values(by="LoanAmount", ascending=False)

#AppClone = App.copy()
AppClone
```

```python
# Grouping

App.groupby("Education").count()

App.groupby("Dependents").count()

App.groupby("Education").mean()

App.groupby("Dependents").std()
 
```

```python
#Plotting

#Plotting from Dataframe

App.plot(kind = 'scatter', x = 'LoanAmount', y = 'ApplicantIncome')

App.plot(kind = 'hist', x = 'LoanAmount', y = 'ApplicantIncome')
```
