# Removing commas to make it meaningful data or the correct format.

```python
countries['Phones'] = countries['Phones'].str.replace(',','')
countries['childdeath'] = countries['childdeath'].str.replace(',','')
countries['Netmgr'] = countries['Netmgr'].str.replace(',','')
countries['Climate'] = countries['Climate'].str.replace('1,5','2') #This is because the data maybe 1.5 actully and can be rounded of i.e to make it 2 that makes some meaning as a part of category
countries['Climate'] = countries['Climate'].str.replace('2,5','3') # same goed for this also.
countries
```

```python
pd.unique(countries['Climate']) # Now we have a meaningful range of category and its datatype will be converted to catergory too.
```

# Replacing commas with '.' to make it in a correct format i.e to a meaningful data 

```python
countries['Pdensity'] = countries['Pdensity'].str.replace(',','.')
countries['Coastline'] = countries['Coastline'].str.replace(',','.')
countries['Arable'] = countries['Arable'].str.replace(',','.')
countries['Crops'] = countries['Crops'].str.replace(',','.')
countries['Other'] = countries['Other'].str.replace(',','.')
countries['Birthrate'] = countries['Birthrate'].str.replace(',','.')
countries['Deathrate'] = countries['Deathrate'].str.replace(',','.')
countries['Agriculture'] = countries['Agriculture'].str.replace(',','.')
countries['Industry'] = countries['Industry'].str.replace(',','.')
countries['Service'] = countries['Service'].str.replace(',','.')
countries['Literacy'] = countries['Literacy'].str.replace(',','.')
countries
```

# conversion of wrong datatype to a correct datatype

```python
countries['Pdensity'] = countries['Pdensity'].astype(float)
countries['Coastline'] = countries['Coastline'].astype(float)
countries['Netmgr'] = countries['Netmgr'].astype(float) #later nedd to convert back to interger after replacing the missing values
countries['childdeath'] = countries['childdeath'].astype(float) #later need to convert back to int after replacing the missing values
countries['Literacy'] = countries['Literacy'].astype(float)
countries['Phones'] = countries['Phones'].astype(float) # convert back to int later after replacing the missing values
countries['Arable'] = countries['Arable'].astype(float) 
countries['Crops'] = countries['Crops'].astype(float) 
countries['Other'] = countries['Other'].astype(float)
countries['Climate'] = countries['Climate'].astype("category") 
countries['Birthrate'] = countries['Birthrate'].astype('float')
countries['Deathrate'] = countries['Deathrate'].astype('float')
countries['Agriculture'] = countries['Agriculture'].astype('float')
countries['Industry'] = countries['Industry'].astype('float')
countries['Service'] = countries['Service'].astype('float')
countries.dtypes
```
